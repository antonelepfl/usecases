First, use module *os* to clean workspace from .gdf or .dat files (file storing spikes and voltage traces from previous simulations, respectively

```python
import os

def remove_files():
    for f in os.listdir('.'):
        if '.gdf' in f or '.dat' in f:
            os.remove(f)
            
remove_files()
```
